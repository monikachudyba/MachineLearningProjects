#Minimum transaction to equalize assets

```python
def equalize (assets):
  if len(assets)<=10**5 and len(assets)>=1:
    if (all(val<=10**9 and val>=0 for val in assets)):
      minTrans = 10**10
      Trans = 0
      for aim in range (max(assets)+1):
        for i in assets:
          Trans += abs(aim-i)
        if Trans<minTrans:
          minTrans = Trans
        Trans= 0
      return minTrans
```

```python
assets = [10, 20, 30, 40, 50, 51, 52, 53, 54, 55]
equalize (assets)
```
